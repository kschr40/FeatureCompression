```python
import numpy as np
import matplotlib.pyplot as plt
import torch
import torch.nn as nn
```

```python
def step_function(x, thresholds):
    return np.sum(x[:,None] > thresholds[None,:], axis=1)

def soft_step_function(x, thresholds, tau):
    x = torch.tensor(x)
    thresholds = torch.tensor(thresholds)
    sigmoid = nn.Sigmoid()
    value = torch.sum(sigmoid((x[:,None] - thresholds[None,:])/tau), axis=1)
    return value.detach().numpy()
```

```python
x = np.linspace(-1,1, 1000)
thresholds = np.array([-0.75, -.1, .4, .75])

y = step_function(x, thresholds)

tau_list = [1, 0.1, 0.06, 0.03, 0.01]
y_soft_dict = {}
for tau in tau_list:
    y_soft_dict[tau] = soft_step_function(x, thresholds, tau)

plt.figure(figsize=(10, 3))    
plt.scatter(x, y)
for i in range(len(thresholds)):
    plt.axvline(thresholds[i], color='r', linestyle='--')
plt.savefig('images/step_function_new.png')
plt.show()
```

```python
plt.figure(figsize=(10, 3))
for tau in tau_list:
    plt.scatter(x, y_soft_dict[tau], label = r'$\tau = {}$'.format(tau), s = 5)
for i in range(len(thresholds)):
    plt.axvline(thresholds[i], color='r', linestyle='--')
plt.scatter(x, y, label = 'Hard quantization function', s = 5) 
plt.legend()
plt.savefig(f'images/quantization_functions_new.png')
plt.show()
```
